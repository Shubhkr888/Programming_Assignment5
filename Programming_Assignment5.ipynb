```python
#Question 2
# 10 - 1,2,5,10
# 15 - 1,3,5,15

def hcf(a,b):
  if a>b:
    number=b
  else:
    number=a
  for i in range(1,number+1):
    if ((a%i==0) and (b%i==0)):
      hcf1 = i
  return hcf1

def hcf2(num1,num2):
  for i in range(1, max(num1, num2)):
    if num1 % i == num2 % i == 0:
          hcf3 = i
  return hcf3
```

```python
hcf(50,100)
```

```python
hcf(15,10)
```
